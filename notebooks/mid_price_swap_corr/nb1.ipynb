```python
import MetaTrader5 as mt5
import pandas as pd
import plotly.express as px
from datetime import datetime, time, timedelta
```

```python
mt5.initialize()
```

```python
ticks = mt5.copy_ticks_range('EURTRY', datetime(2023, 9, 1), datetime.now() + timedelta(hours=4), mt5.COPY_TICKS_ALL)
ticks = pd.DataFrame(ticks)
ticks['time'] = pd.to_datetime(ticks['time_msc'], unit='ms')
ticks['mid_price'] = (ticks['bid'] + ticks['ask']) / 2
ticks['time2'] = ticks['time'].apply(datetime.time)

ticks['mid_price_diff'] = ticks['mid_price'] - ticks['mid_price'].shift(1)
ticks
```

```python
cond1 = ticks['time2'] > time(23, 30, 0)
cond2 = ticks['time2'] < time(0, 30, 0)

ticks2 = ticks[cond1 | cond2].copy()
ticks2
```

```python
px.scatter(ticks2, x='time', y='mid_price')
```

```python
ticks2['arb'] = ticks2['bid'] - ticks2['ask'].shift(-1)
ticks2['time_diff'] = ticks2['time'].shift(-1) - ticks2['time']
ticks2
```

```python
cond1 = ticks2['arb'] > 0
cond2 = ticks2['time_diff'] < timedelta(minutes=1)

ticks2['day'] = ticks2['time'].dt.dayofweek

ticks2[cond1 & cond2].sort_values('arb', ascending=False)
```
